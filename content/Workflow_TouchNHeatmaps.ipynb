```python
from PIL import Image
```

```python
pitch = Image.open("emirates_pitch.png")
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
raw_data = pd.read_csv("draft.csv")
```

```python
df = raw_data.copy()
```

```python
player = "elneny"
```

```python
touches_received = df.loc[ (df["player2"] == player) & (df["event"] == "completed_pass")  ,  ["end_x", "end_y"]].copy()
```

```python
touches_received.head()
```

```python
touches_received.isna().values.any()
```

```python
fig, ax = plt.subplots(figsize=(52.5, 34))
plt.axis( [ 0, 105, 0, 68])
ax.imshow(pitch, extent=[0,105,0,68])

plt.scatter(x = touches_received["end_x"], y = touches_received["end_y"], s=1000)
plt.tight_layout()
#c="cornflowerblue", edgecolors="gold", linewidth=5, alpha=0.8)
```

```python
fig, ax = plt.subplots(figsize=(52.5, 34))
plt.axis( [0, 105, 0, 68] )
ax.imshow(pitch, extent=[0,105,0,68])

sns.kdeplot( x = touches_received["end_x"], y = touches_received["end_y"],
     n_levels=30, cmap="rocket_r", fill=True, alpha=0.8)
plt.tight_layout()
#plt.savefig("BBK_elnenyheatwip.png"); #"rocket", "mako", "flare", and "crest". 
```
